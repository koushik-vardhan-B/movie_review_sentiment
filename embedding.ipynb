```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
## Sentences
sent = [
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am a good boy',
    'I am a good developer',
    'understand the meaning of words',
    'your videos are good'
]
```

```python
sent
```

```python
## Define the vocabulary size
voc_size = 10000
```

```python
## one_hot representation
one_hot_repr = [one_hot(words,voc_size) for words in sent]
```

```python
one_hot_repr
```

```python
## Word embedding representation

from tensorflow.keras.layers import Embedding
# from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length = 8
embedded_docs = pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
## feature representation
dim = 10
model = Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length,trainable=True))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```

```python
## feature representation
# Predict on a single sample, reshaping it to include a batch dimension
prediction = model.predict(np.expand_dims(embedded_docs[0], axis=0))
print(prediction)
```
